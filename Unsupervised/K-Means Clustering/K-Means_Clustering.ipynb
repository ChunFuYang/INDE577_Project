# K-Means Clustering
This notebook demonstrates K-Means Clustering on a load_wine dataset.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

## Load and Prepare the Data

```python
wine = load_wine()
X = wine.data
y = wine.target
n_clusters = len(np.unique(y))
print(f"Feature names: {wine.feature_names}")
print(f"Number of classes: {n_clusters}")
```

```python
print(f'Number of classes: {len(np.unique(y))}')
```

## Data preprocessing

```python
X = StandardScaler().fit_transform(X)
```

## Apply K-Means Clustering

```python
kmeans = KMeans(n_clusters=7, random_state=42)
clusters = kmeans.fit_predict(X)
```

## Visualize data with PCA

```python
pca = PCA(n_components=13)
X_pca = pca.fit_transform(X)
pca.fit(X_pca)
X_projected = pca.transform(X_pca)

plt.figure(figsize=(8,6))
# Plotting the clusters
plt.scatter(X_projected[:,0], X_projected[:,1], c=y, cmap='viridis', alpha=0.8)

plt.title('Original data labels using PCA')
plt.colorbar(label='Cluster Label')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

## Visualize results using PCA

```python
#training the KMeans model
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
# predict the cluster labels
y_kmeans = kmeans.predict(X)
# plot the clusters of load_wine dataset using pca
pca = PCA(n_components=13)
pca.fit(X)
X_projected = pca.transform(X)

plt.figure(figsize=(8,6))
# plot the clusters
plt.scatter(X_projected[:,0], X_projected[:,1], c=y_kmeans, cmap='viridis', alpha=0.8)
centers = kmeans.cluster_centers_
#project the cluster centers to the PCA space
centers = pca.transform(centers)
#add the cluster centers to the plot
plt.scatter(centers[:,0], centers[:,1], c='red', s=200, alpha=0.75, marker='x')
plt.title('K-Means Clustering Results using PCA')
plt.colorbar(label='Cluster Label')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.show()
```

## Analysis and Discussion
Even though K mean clustering is a unsupervised that works on unlabeled data. I used the original label to visual how the origrinal data clusters look like so that I can analysis how well KNN works. There is no clear boundary between classes in the original data, but if we see the results of the KNN, we can see the boundary of the clusters can be easy distinguished. Since KNN is only computing Euclidean distance between data points, there are some details of the feature that can be captures. The KNN with 3 clusters actually did quite decent on clustering which was similar to original labesl. However, when the K gets to large e.g 7, it becomes hard to distinguish the boundary.
